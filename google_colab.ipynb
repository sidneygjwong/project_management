## Using R in Google Colab

If you're unfamiliar with Python, that's okay! You also run Google Colab with R or Julia by changing the Runtime. If you want to run R code without changing the Runtime, you can run R by activating R magic by running the following code:

```
%load_ext rpy2.ipython
```

Try running the code below:

````python
# activate R magic
%load_ext rpy2.ipython
````

Now you can run R in your Google Colab by prefixing your code chunk with ```%%R```.

````python
%%R
n <- 1 + 1
n
````

Cool! You'll notice the Red squiggly line to indicate an error in your code, but it's only because it's trying to interpret your code in Python. Let's try another example:

````python
%%R
m <- "Hello"
print(m)
````

## Integrating with Google Drive

One benefit of Google Colab is that you can integrate it with your Google Drive. You will first need to mount your personal Google Drive account using the following code.

```
from google.colab import drive
drive.mount('/content/drive')
```

Here's a live example:

````python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')
````

